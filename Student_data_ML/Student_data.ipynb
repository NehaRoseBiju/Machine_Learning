The objective of this task is to perform end-to-end data preprocessing on the Student
Dataset using Google Colab. This includes dataset exploration, handling missing values, encoding categorical variables, and preparing the dataset for machine learning.
The goal is to ensure the dataset is clean, structured, and ready for model training.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
#Load the CSV file using Pandas
df=pd.read_csv('/content/student_data_100_preprocess.csv')
```

```python
#Display:– First 5 rows– Last 5 rows
print("first 5 rows : ")
print(df.head())
```

```python
print("last 5 rows : ")
print(df.tail())
```

```python
#– Shape of the dataset
print("Shape :")
print(df.shape)
```

```python
#– Column names
print("column names:")
print(df.columns.tolist())
```

```python
#– Data types
print("data types:")
print(df.dtypes)
```

```python
#– Numerical columns
num_col = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("numerical columns:")
print(num_col)
```

```python
#– Categorical columns
cat_col = df.select_dtypes(include=['object', 'category']).columns.tolist()
print("categorical columns:")
print(cat_col)
```

```python
#statistical summary of numerical columns
print("statistical summary:")
print(df[num_col].describe())
```

Key statistical measures including mean, standard deviation, minimum, maximum, and quartile values for each numerical column.

```python
#count missing values per column and Identify which columns contain missing data
df.isnull().sum()
```

```python
#– Mean filling
for col in num_col:
    if df[col].isnull().sum() > 0:
        mean_value = df[col].mean()
        df[col].fillna(mean_value, inplace=True)
        print(f"Filled missing values in '{col}' with Mean = {mean_value:.2f}")
```

```python
#– Mode filling
for col in cat_col:
    if df[col].isnull().sum() > 0:
        mode_value = df[col].mode()[0]
        df[col].fillna(mode_value, inplace=True)
        print(f"Filled missing values in '{col}' with Mode = {mode_value}")
```

```python
#Verify:– No missing values remain
print("missing values :")
print(df.isnull().sum())
```

```python
#confirm changes
df.info()
```

```python
df.describe()
```

```python
#Encode Gender using Label Encoding
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
```

```python
#Apply One-Hot Encoding to Parental_Education using Pandas
df = pd.get_dummies(df, columns=['Parental_Education'])
```

```python
df.info()
```

```python
#Remove original categorical columns
df.drop('Gender', axis=1, inplace=True)
```

```python
df.info()
```

```python
# Separate:– Features (X)– Target (y = Final_Score)
X = df.drop('Final_Score', axis=1)  # Features
y = df['Final_Score']               # Target
print("features shape:", X.shape)
print("target shape:", y.shape)
```

```python
# Split the dataset:– 80% training and 20% testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

In this experiment, the student dataset was successfully preprocessed. The dataset was explored to understand its structure, missing values were identified and handled using appropriate imputation techniques, categorical variables were encoded into numerical formats, and the data was prepared for machine learning by separating features and target variables and performing an 80-20 train-test split. The resulting dataset is clean, consistent, and suitable for predictive modeling tasks.
