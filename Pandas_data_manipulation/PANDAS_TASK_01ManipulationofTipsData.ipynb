```python
#9.Feature Creation
#Tip Percentage
tips["tip_percentage"]=(tips["tip"]/tips["total_bill"])*100
```

```python
tips.head()
```

```python
#Binning
tips["Bill_Category"]=pd.cut(
    tips["total_bill"],
    bins=[0,10,20,30,60],
    labels=["Low","Medium","High","Very High"]
)
```

```python
tips.head()
```

The new feature gives us the information about weather a customer bill is high ,low or medium.

```python
#10.GroupBy
#Gender
tips.groupby("sex").mean(numeric_only=True)
```

```python
#Day
tips.groupby("day").mean(numeric_only=True)
```

```python
#11.Multi-Level Grouping
tips.groupby(["day","sex"]).mean(numeric_only=True)
```

Spending varies according to both gender and day.

```python
#12.Crosstab
pd.crosstab(tips["day"],tips["time"])
```

Dinner is more common than lunch. Especially during weekends people come more frequently to eat dinner.

```python
#13.Pivot Table
pd.pivot_table(
    tips,
    values="tip",
    index="day",
    columns="sex",
    aggfunc="mean"
)
```

Male customers leave higher average tips than female customers on both Saturday and Thursday, while female customers leave slightly higher average tips on Friday and Sunday.
Sunday has the highest average tip for both genders, indicating that customers tend to tip more on that day.
The differences in average tips between males and females are relatively small, suggesting that the day of the visit has a greater influence on tipping than gender.

```python
#14.Ranking
#Bill Rank
tips["Bill_Rank"]=tips["total_bill"].rank(ascending=False)
```

```python
#Tip Percentage Rank
tips["Tip_Rank"]=tips["tip_percentage"].rank(ascending=False)
```

```python
tips.head()
```

Customers with the same bill amount or tip percentage receive the same average rank (for example, ranks such as 82.5 or 219.5), showing that the rank() function assigns tied values the average of their positions.Customers with higher total bills generally receive better (lower-numbered) bill ranks, making it easy to identify the restaurant's highest spenders.
The Tip_Rank is based on tip percentage rather than tip amount, so customers with smaller bills can still achieve a high rank if they leave a generous percentage as a tip.
A customer with a high Bill_Rank does not always have a high Tip_Rank, indicating that spending more does not necessarily mean tipping a higher percentage.

```python
#19.Visualization
#bar chart
tips.groupby("day")["Bill"].mean().plot(kind="bar")
plt.title("Average Bill by Day")
plt.show()
```

Sunday has the highest average bill, indicating that customers spend the most on this day.
Friday has the lowest average bill among all four days, suggesting that customers spend less compared to the rest of the week.
Weekend days (Saturday and Sunday) have higher average bills than weekdays (Thursday and Friday), indicating increased customer spending during weekends.

```python
#pie chart
tips["sex"].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.ylabel("")
plt.show()
```

Male customers make up the majority of the restaurant's customers, accounting for approximately 64.6% of the dataset, while female customers account for about 35.4%.
The proportion of male customers is almost twice that of female customers, indicating that the dataset contains significantly more male dining records.
The gender distribution is not evenly balanced, which should be considered when comparing customer behavior, spending patterns, or tipping habits between males and females.

```python
#line chart
tips["Bill"].head(30).plot(kind="line")
plt.title("Bills of First 30 Customers")
plt.show()
```

The total bill amounts fluctuate considerably among the first 30 customers, indicating that customer spending varies from one visit to another.
A few customers have exceptionally high bills (around 35–40), while most customers have bills between approximately 10 and 25, showing that high-value transactions are less common.
There is no consistent increasing or decreasing trend in the bill amounts, suggesting that customer spending is irregular and depends on individual dining behavior rather than the order of transactions.

```python
#Scatter Plot
tips.plot(
    kind="scatter",
    x="Bill",
    y="Tip"
)
plt.show()
```

There is a positive relationship between the total bill and the tip amount, indicating that customers with higher bills generally leave higher tips.
Most customer bills fall between approximately 10 and 30, with corresponding tips ranging from about 1 to 5, showing that the majority of transactions are in the moderate spending range.
Although the overall trend is positive, customers with similar bill amounts sometimes leave different tip amounts, suggesting that tipping behavior varies from customer to customer.

```python
#Stacked Bar Chart
pd.crosstab(
    tips["day"],
    tips["sex"]
).plot(kind="bar",stacked=True)
plt.show()
```

Saturday has the highest total number of customers, while Friday has the lowest customer count among all four days.
Male customers outnumber female customers on Friday, Saturday, and Sunday, whereas Thursday has more female customers than male customers.
Weekend days (Saturday and Sunday) attract more customers overall than weekdays, indicating that the restaurant is busier during the weekend.
