Through this tadk we can understand how  different imputation methods influence data distribution and support
appropriate method selection during data preprocessing.Imputation is the act of assigning a convinient value to the whole data without changing the meaning of the data.The selection of imputation methods depend on the type of data, existence of outliers etc.

```python
import pandas as pd
import numpy as np
```

```python
# Creating the dataset
data = {
    'Age': [25, 30, np.nan, 45, 50, np.nan, 35],
    'Blood_Pressure': [120, 125, 130, np.nan, 140, 135, np.nan],
    'Cholesterol': [200, np.nan, 180, 220, np.nan, 210, 190],
    'Gender': ['Male', 'Female', np.nan, 'Female', 'Male', np.nan, 'Female']
}
df = pd.DataFrame(data)

print("Original Dataset:")
display(df)
```

```python
#Inspect Dataset
#Dataset Information
df.info()
```

The database contains the columns age,blood pressure ,cholesterol and gender. Gender contains categorical value while all teh other columns have numerical values.

```python
#Summary Statistics
df.describe()
```

```python
#missing values count
print(df.isnull().sum())
```

We use different imputations to replace the missing values present in the dataset.

```python
#mean imputation
df_mean = df.copy()
numeric_cols = ['Age', 'Blood_Pressure', 'Cholesterol']
for col in numeric_cols:
    df_mean[col] = df_mean[col].fillna(df_mean[col].mean())
print("Dataset After Mean Imputation")
display(df_mean)
```

Using mean imputation we can replace the missing values with the arithemetic value of the column. It does not change the overall average and can help to reduce variance. But it can distort data if there is existence of outliers. It may weaken the relationship between variables and may not be suitable for highly skewed data.It is generally used for numerical data and is simple and fast.

```python
df_mean.describe()
```

The missing vavlues in the numerical columns( age,blood pressure and cholestrol) are replaced with the mean value of the column)+.

```python
#median imputation
df_median = df.copy()
for col in numeric_cols:
    df_median[col] = df_median[col].fillna(df_median[col].median())
print("Dataset After Median Imputation")
display(df_median)
```

In median imputation the original values are sorted and the median is calculated which is later used to replace the missing values.It is an imputation method that is resistant to outliers and is better choice for skewed distributions. In comparison it preserves the central tendancy more acurately.

```python
df_median.describe()
```

The missing vavlues in the numerical columns( age,blood pressure and cholestrol) are replaced with the median value of the column.

```python
#mode imputation
df_mode = df.copy()
gender_mode = df_mode['Gender'].mode()[0]
df_mode['Gender'] = df_mode['Gender'].fillna(gender_mode)
print("Dataset After Mode Imputation")
display(df_mode)
```

In mode imputation the missing values are replaced with the most frequently occuring values. It is used when the column comprises of categorical values .It leads to an increse in the majority values of the feature hence leads to an increase in the majority class bias.

```python
print("Before mode imputation")
print(df['Gender'].value_counts(dropna=False))
print("\nAfter mode imputation")
print(df_mode['Gender'].value_counts(dropna=False))
```

usin mode imputation we replace the missing value in the categorical column Gender. The missing value is replaced with the majority gender here which is female

```python
#comparison of mean and median imputation
print("Original Dataset Statistics")
display(df.describe())

print("Mean Imputation Statistics")
display(df_mean.describe())

print("Median Imputation Statistics")
display(df_median.describe())
```

Mean imputation is used when there is numerical values and no significiant outliers.Median imputation is used when the numerical data contain outliers.Mode imputation is used when the values are categorical in nature. In above datasets which comes under healthcare sector we generally make use of the median imputation because medical values often consists of extreme values.
