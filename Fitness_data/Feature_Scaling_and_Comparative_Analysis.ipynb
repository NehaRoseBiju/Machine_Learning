```python
# Import required libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# Load dataset
df = pd.read_csv('/content/fitness_activity_100+_preprocess (1).csv')
```

The dataset has been loaded successfully.

```python
# Display first few rows
df.head()
```

```python
# Dataset information
print(df.info())
```

The column User_ID is of the type object which all other columns are numerical.From the statistical summary we obtained the mean standard deviationa etc.


```python
# Statistical summary
print(df.describe())
```

```python
#Select Features for Scaling
features = ['Steps_Per_Day', 'Calories_Burned', 'Workout_Minutes']
X = df[features]
```

```python
# Standard Scaling
standard_scaler = StandardScaler()
X_standard = standard_scaler.fit_transform(X)
standard_df = pd.DataFrame(
    X_standard,
    columns=features
)
print("Standard Scaled Data")
standard_df.head()
```

```python
# Min-Max Scaling
minmax_scaler = MinMaxScaler()
X_minmax = minmax_scaler.fit_transform(X)
minmax_df = pd.DataFrame(
    X_minmax,
    columns=features
)
print("Min-Max Scaled Data")
minmax_df.head()
```

```python
compare = pd.concat(
    [
        X.head(),
        standard_df.head(),
        minmax_df.head()
    ],
    axis=1
)
compare
```

```python
#Statistical Comparison
print("Standard scaler statistics")
print(standard_df.describe())
print("\nmin-max scaler statistics")
print(minmax_df.describe())
```

The features were transformed using StandardScaler to have mean as 0 and standard deviation as 1. It retains the effect of outliers. The Minmaxscaler converted all the values to the range 0 to 1. It copressed the values so that it lies between a selected range which is 0 to 1. Both of these techniques turned out to be effective for scaling.
